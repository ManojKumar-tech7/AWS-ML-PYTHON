```python
# function calculator
def calculator(num1,num2,op):
  if op=="+":
    return num1+num2
  elif op=="-":
    return num1-num2
  elif op=="*":
    return num1*num2
  elif op=="/":
    return num1/num2
  else:
    print("Invalid operator")

result = calculator(10,20,"+")
print(result)
```

```python
def cal(*a):
  print(eval(f"{a[0]} {a[1]} {a[2]}"))
cal(10,"+",20)
cal(10,"-",20)
cal(10,"*",20)
cal(10,"/",20)
cal(10,"%",20)
```
