*Operators*
1. arith -> + , - , * , / , // , % , ** , etc
2. relational -> < , > , >= , <= , == , != , etc
3. assignment -> += , -= , *= , /= , etc
4. logical -> and , or , not , etc
5. bitwise -> & , |

```python
# Arithmetic operators
# 1. addition operator
num1 = 13
num2 = 20
print(num1+num2) #33
```

```python
# 2. subtraction operator
num1 = 100
num2 = 57
print(num1-num2) # 43
```

```python
# 3. multiplication operator
num1 = 7
num2 = 3
print(num1*num2) # 21
```

```python
# 4. division operator
num1 = 21
num2 = 3
print(num1/num2) # 7.0
```

```python
# 5. floor division operator
num1 = 50
num2 = 3
print(num1 // num2) # 16 -> Quotient
```

```python
# 6. power operator
num1 = 2
num2 = 5
print(num1 ** num2) #32
```

```python
# Relational operator
# 1. < (less than operator)
num1 = 17
num2 = 57
print(num1<num2) # True (since num1 less than num2)
```

```python
# 2. > (greater than operator)
num1 = 34
num2 = 87
print(num1>num2) # False (since num1 is not greater than num2)
```

```python
# 3. <= (less than equala to operator)
num1 = 51
num2 = 52
print (num1<=num2) # True (since num1 is less than num2)
num3 = 50
num4 = 50
print(num3<=num4) # True (since num3 is equals to num4)
```

```python
# 4. >= (greater than equals to operator)
num1 = 51
num2 = 52
print (num1>=num2) # False (since num1 is not greator than num2)
num3 = 50
num4 = 50
print(num3<=num4) # True (since num3 is equals to num4)
```

```python
# 5. == (equals to operator)
num1 = 50
num2 = 50
print(num1<=num2) # True (since num1 is equals to num2)
```

```python
# 6. != (not equals to operator)
num1 = 33
num2 = 33
print(num1!=num2) # False (since num1 is not equals to num2)
```

```python
# Assignment operator
# 1. += operator
num1 = 22
num2 = 33
num1+=num2
print(num1) # 55
```

```python
# 2. -= operator
num1 = 33
num2 = 22
num1-=num2
print(num1) # 11
```

```python
# 3. *= operator
num1 = 2
num2 = 5
num1*=num2
print(num1) # 10
```

```python
# 4. /= operator
num1 = 8
num2 = 2
num1/=num2
print(num1) # 4.0
```

```python
# logical operator
# 1. and operator
num1 = 10
num2 = 20
print(num1 and num2) # 20

name = "Hello"
num3 = 3
print(name and num3) # 3
print(num3 and name) # Hello
```

```python
# 2. or operator
num1 = 22
num2 = 67
print(num1 or num2) # 22
print(num2 or num1) # 67

str = "world"

print(num1 or str) # 22
print(str or num2) # world
```

```python
# 3. not operator
num1 = True
print(num1) # True
print(not(num1)) # False
```
