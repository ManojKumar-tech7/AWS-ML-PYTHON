*Operators*
1. arith -> + , - , * , / , // , % , ** , etc
2. relational -> < , > , >= , <= , == , != , etc
3. assignment -> += , -= , *= , %= , etc
4. logical -> and , or , not , etc
5. bitwise -> & , |

```python
# Arithmetic operators
# 1. addition
num1 = 13
num2 = 20
print(num1+num2) #33
```

```python
# 2. subtraction
num1 = 100
num2 = 57
print(num1-num2) # 43
```

```python
# 3. multiplication
num1 = 7
num2 = 3
print(num1*num2) # 21
```

```python
# 4. division
num1 = 21
num2 = 3
print(num1/num2) # 7.0
```

```python
# 5. floor division
num1 = 50
num2 = 3
print(num1 // num2) # 16 -> Quotient
```

```python
# 6. power
num1 = 2
num2 = 5
print(num1 ** num2) #32
```

```python
# Relational operator
# 1. < (less than)
num1 = 17
num2 = 57
print(num1<num2) # True (since num1 less than num2)
```

```python
# 2. > (greater than)
num1 = 34
num2 = 87
print(num1>num2) # False (since num1 is not greater than num2)
```

```python
# 3. <= (less than equala to)
num1 = 51
num2 = 52
print (num1<=num2) # True (since num1 is less than num2)
num3 = 50
num4 = 50
print(num3<=num4) # True (since num3 is equals to num4)
```

```python
# 4. >= (greater than equals to)
num1 = 51
num2 = 52
print (num1>=num2) # False (since num1 is not greator than num2)
num3 = 50
num4 = 50
print(num3<=num4) # True (since num3 is equals to num4)
```
