```python
class User1:
  def __init__(self,pen,pencil):
    print("User1 constructor is working")
    self.pen = pen
    self.pencil = pencil
  def penGet(self):
    print("pen is working: ",self.pen)
  def pencilGet(self):
    print("pencil is working: ",self.pencil)

class User2(User1):
  def __init__(self,pen,pencil,earser):
    print("User2 constroctor is working")
    super().__init__(pen,pencil)
    self.earser = earser
# object
u = User2("blue","black","doms")
u.penGet()
```

```python
# encapsulation
class Student:
  def __init__(self,name,roll_no):
    self.__data = name
    self.roll_no = roll_no
  def getname(self,password = None):
    if password == "123456":
      print(self.__data)
    else:
      print("Incorrect password")


s = Student("Manoj",73)
print(s.roll_no)
# print(s.__data) -> show error of object has no attribute '__data'
s.getname("123456")
```
