```python
# operations on columns
import pandas as pd
data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
df = pd.DataFrame(data)

# drop cols
df.drop("Emp_ID",axis=1,inplace=True)
print(df)

# drop multi cols
df.drop(["Name","Department"],axis=1,inplace=True)
print(df)
```

```python
# operations on rows
import pandas as pd
data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
df = pd.DataFrame(data)
print(df)

# loc 1-> index of row , second -> cols name
print(df.loc[1,"Name"])

# iloc row index -> 0 and cols index -> 0
print(df.iloc[1,1])
```

```python
# filter
import pandas as pd
data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
df = pd.DataFrame(data)
print(df)
# filter one greater
print(df[df['Emp_ID'] > 104])

# 106
print(df[df['Emp_ID'] == 106])

# 30000 to 50000
# print(df[(condition one) & (condition 2)])
print(df[(df['Salary']> 30000) & (df['Salary'] < 50000)])
```

```python
# sorting
import pandas as pd
data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
df = pd.DataFrame(data)
print(df)

# sorting ascending
print(df.sort_values("Salary",ascending=False))

# two cols
print(df.sort_values(["Salary","Experience"]))
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/rajendra0968jangid/Ds-Arya/main/student-data.json"
df = pd.read_json(url)
print(df)
```

```python
print(df.head(5))
print(df.tail(1))
```

```python
print(df.shape)
```

```python
# Filter rows where gender is Female
female_students = df[df['gender'] == 'Female']
print(female_students)
```

```python
# Filter by a list of names
specific_students = df[df['name'].isin(['arun', 'usha', 'Jenifer'])]
print(specific_students)
```

```python
# Use &
top_math_males = df[(df['maths'] > 80) & (df['gender'] == 'Male')]
print(top_math_males)
```

```python
#Sorting!!
sorted_by_physics = df.sort_values(by='physics', ascending=False)
print(sorted_by_physics)
```

```python
# multiple sort
multi_sorted = df.sort_values(by=['gender', 'maths'], ascending=[True, False])
print(multi_sorted)
```

```python
# combined filter and sort for testing purpose
# Filter for English > 70, then sort by English descending
high_english = df[df['english'] > 70].sort_values(by='english', ascending=False)
print(high_english)
```
