```python
# inheritance
class Student:
  def __init__(self,name,roll_no):
    self.name = name
    self.roll_no = roll_no
  def getData(self):
    print(self.name,self.roll_no)

class Address(Student):
  def __init__(self,name,roll_no,city,state):
    #self.name=name
    #self.roll_no=roll_no
    super().__init__(name,roll_no)
    self.city = city
    self.state = state
  def getAddress(self):
    print(self.city,self.state)


a = Address("manoj","23","jaipur","rajasthan")
a.getData()
a.getAddress()
```
